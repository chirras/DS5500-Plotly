```python
# Filtering the data to get records of 'AMZN' stock.

df_amz = df_price[df_price.symbol == 'AMZN'].reset_index(drop=True)
df_amz.head()
```

```python
# Formatting the date

df_amz.date = pd.to_datetime(df_amz.date, format='%Y-%m-%d')
```

```python
# Creating OHLC Plot

trace = go.Ohlc(x = df_amz.date,
                open = df_amz.open, 
                high = df_amz.high, 
                low = df_amz.low, 
                close = df_amz.close)

data = [trace]
layout = {
    'title': 'Price movement of Amazon stock using OHLC Plot',
    'yaxis': {'title': 'AMZN Stock Price'},
    'xaxis': {'title': 'Years',
             'rangeslider': {'visible': False}}
}

fig = dict(data=data, layout=layout)

py.iplot(fig, validate=False)
```

Describe and justify your visual encoding / design choices (2 sentences)

    The above OHLC plot uses 'Lines' as the marks and 'Position' as the channel. As position is top ranked in channel ranking by datatype for both the datatypes, this plot is one of the effective way to visualize the given data. It uses Color to indicate the price movement of the stock i.e., 'Green' indicates increase in price and 'Red' indicates decrease in price. It also uses small horizontal lines to indicate the open and close prices for a day.

```python

fig = ff.create_candlestick(open = df_amz.open, 
                            high = df_amz.high, 
                            low = df_amz.low, 
                            close = df_amz.close, 
                            dates=df_amz.date)

fig['layout'].update({
    'title': 'Price movement of Amazon stock using Candlestick Plot',
    'xaxis': {'title': 'Years'},
    'yaxis': {'title': 'AMZN Stock Price'}
})


py.iplot(fig, filename='finance/aapl-candlestick', validate=False)
```

If the upper shadow is short, it indicates that the open that day was near the high of the day and viceversa. The relationship between the day's open, high, low and close determines the look of candlestick
